```python
from scipy.interpolate import interpn
from helpFunctions import surfacePlot
from functools import partial
import numpy as np
def value_func_3d(x, y):
    return 2 * x + 3 * y
```

```python
x = np.array([1,2,3])
y = np.array([2,4,6,8])
points = (x,y)
xx, yy = np.meshgrid(*points, indexing='ij')
print(xx)
print(yy)
print(xx.shape)
```

```python
values = value_func_3d(*np.meshgrid(*points, indexing='ij'))
values
```

```python
pX, pY = np.meshgrid(*(np.linspace(0,10,11), np.linspace(0,10,11)))
```

```python
point = np.array(list(zip(pX.flatten(), pY.flatten())))
```

```python
values.shape
```

```python
L =[]
for p in point:
    L.append(interpn(points, values, p, bounds_error = False, fill_value = None)[0])
L = np.array(L)
L = L.reshape((len(pX),len(pY)))
```

```python
surfacePlot(xx, yy, values)
```

```python
surfacePlot(pX, pY, L)
```
